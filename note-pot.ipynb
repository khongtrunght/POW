```python
# Test triangle inequality
for i in range(10000):
    x = np.random.rand(10)
    y = np.random.rand(10)
    z = np.random.rand(10)
    M_xy = ot.dist(x.reshape(-1,1), y.reshape(-1,1), metric="euclidean")
    M_yz = ot.dist(y.reshape(-1,1), z.reshape(-1,1), metric="euclidean")
    M_xz = ot.dist(x.reshape(-1,1), z.reshape(-1,1), metric="euclidean")

    s = 0.8

    p = ot.unif(x.shape[0])
    q = ot.unif(y.shape[0])
    r = ot.unif(z.shape[0])
    
    w_xy = partial_wasserstein2(p,q,M_xy,m=s,log=False)
    w_yz = partial_wasserstein2(q,r,M_yz,m=s,log=False)
    w_xz = partial_wasserstein2(p,r,M_xz,m=s,log=False)

    if w_xz > w_xy + w_yz:
        print("Not satisfied")

    if w_xy > w_xz + w_yz:
        print("Not satisfied")

    if w_yz > w_xz + w_xy:
        print("Not satisfied")
```
